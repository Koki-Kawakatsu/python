```python
import pandas as pd

df = pd.read_csv('Ikey240611005A4-wave.csv', header=None)
column_names = ["Column1", "Column2"] 

df.columns = column_names

#print(df.head())
df
```

```python
df['Column2'] = -df['Column2']
```

```python
df.plot(0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = df['Column1']
y = df['Column2']

# Plot the waveforms
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

ax.plot(x, y, color='black', linewidth=2)
ax.axvline(x=2500, color='red', linestyle='--', linewidth=2, label='2500ms')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('Time (ms)', fontsize=20)
ax.set_ylabel('Î”F/F0 (%)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20, width=2)

plt.tight_layout()
plt.show()
```
